## Research Question

To what extent are age, sex, race/ethnicity, and socioeconomic status predictive of disparities in sleep duration (hours per night) and sleep-related impairment (self-reported daytime functional limitations), after adjusting for self-reported mental health status, among U.S. adults?

## Background and Prior Work

Sleep is a core factor in overall health and daily functioning, and public health guidance typically advises <7 hours/night as insufficient sleep for adults.<sup><a href="#ref1" id="cite1">1</a></sup>

National surveillance data show that insufficient sleep is common in the U.S. and also that insufficient sleep differs across demographic groups.
For example, the CDC analysis of the Behavioral Risk Factor Surveillance System found that only about two-thirds of U.S. adults reported “healthy sleep duration” (≥7 hours), with a lower rate among several racial/ethnic minority groups and among adults without higher education, this shows that there is broader concern for the sleep-health of the overall population rather than individual sleep choices.<sup><a href="#ref2" id="cite2">2</a></sup>
Therefore, sleep duration is a population-level concern that could also potentially contribute to chronic disease and impaired cognition.<sup><a href="#ref2" id="cite2b">2</a></sup>

A research paper by the National Library of Medicine has investigated why sleep duration varies by age, sex, race/ethnicity, and socioeconomic status. The medical reviews emphasize that “race/ethnicity” is a social category that can include potential exposures—such as discrimination, poor neighborhood conditions, and working constraints—that ultimately play a role in sleep opportunity and timing.<sup><a href="#ref3" id="cite3">3</a></sup>
Large U.S. survey studies show that disparities stay even after accounting for many variables. For instance, an NHIS trends analysis reported increases in self-reported short sleep in recent years that were especially pronounced among Black and Hispanic adults even after adjusting for demographic and socioeconomic factors, raising concern that widening sleep disparities may lead to other health inequalities over time.<sup><a href="#ref4" id="cite4">4</a></sup>
Longitudinal evidence also suggests that short sleep is more common in lower income/education groups and among some racial/ethnic minority groups, with socioeconomic context and living conditions explaining more of the disparity than individual health behaviors alone.<sup><a href="#ref5" id="cite5">5</a></sup>

Importantly, the consequences of insufficient sleep extend beyond “hours slept” to sleep-related impairment—fatigue, excessive daytime sleepiness, and reduced functioning.
Research that took into account both sleep duration and symptoms found that lack of sleep is associated with higher levels of fatigue and daytime sleepiness, and that both education and race/ethnicity relate to multiple sleep-related symptoms, underscoring that disparities may appear in both sleep quantity and daytime impact.<sup><a href="#ref6" id="cite6">6</a></sup>
Additionally, Mental health is central to this question because it is strongly connected with overall sleep health: evidence supports relationships between sleep disturbances and depression/anxiety, and anxiety can also cause excessive daytime sleepines.<sup><a href="#ref7" id="cite7">7</a></sup>
Also, consistent with that, NHIS-based analysis showed that psychological distress is directly associated with unhealthy sleep duration even after adjusting for socioeconomic factors and health conditions.<sup><a href="#ref8" id="cite8">8</a></sup>
Together, the research shows the importance of our project's focus on estimating how age, sex, race/ethnicity, and socioeconomic status relate to both sleep duration and sleep-related impairment.

## Hypothesis


We hypothesize that age, sex, race/ethnicity, and socioeconomic status will be significantly associated with sleep duration and sleep-related impairment among U.S. adults, even after adjusting for self-reported mental health status. Specifically, individuals from lower socioeconomic backgrounds and racial/ethnic minority groups are expected to report shorter sleep duration and greater sleep-related impairment. This hypothesis is based on prior research showing that social and structural factors influence both sleep health and access to resources that support healthy sleep.


According to prior information, studies have consistently shown that racial and ethnic minority groups and individuals with lower socioeconomic status report shorter sleep duration and greater sleep-related impairment compared to more advantaged groups, even when accounting for individual health behaviors. These disparities are often attributed to structural factors such as nonstandard work schedules, financial strain, neighborhood conditions, chronic stress, and reduced access to healthcare, all of which can negatively affect sleep patterns and daytime functioning. Although mental health is closely linked to sleep outcomes, it does not fully explain observed differences in sleep across demographic groups. Therefore, adjusting for self-reported mental health status is expected to reduce but not eliminate disparities in sleep duration and sleep-related impairment, suggesting an independent role of age, sex, race/ethnicity, and socioeconomic status in shaping sleep health among U.S. adults.


## Data

The ideal dataset to answer this research question would include a large, nationally representative sample of U.S. adults with detailed information on sleep, demographic characteristics, socioeconomic status, and mental health. Key outcome variables would include sleep duration measured in hours per night and indicators of sleep-related impairment, such as self-reported daytime functional limitations due to poor sleep. Primary independent variables would include age, sex, race/ethnicity, and socioeconomic status, measured through indicators such as income and educational attainment, while self-reported mental health status would serve as an important control variable.

A large number of observations would be needed to reliably estimate disparities across demographic subgroups, particularly for racial/ethnic and socioeconomic comparisons. Ideally, the dataset would include thousands of respondents to ensure sufficient data is collected and that it is representative of the entire population. These data would be collected through standardized population-based surveys sdults across the United States using consistent questionnaires. The dataset would be stored in a structured, tabular format like CSV or SAS, with each row representing an individual respondent and each column representing a variable, allowing for efficient data cleaning and statistical analysis in Python.
