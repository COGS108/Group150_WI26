# Research Question

To what extent are age, sex, race/ethnicity, and socioeconomic status predictive of disparities in sleep duration (hours per night) and sleep-related impairment (self-reported daytime functional limitations), after adjusting for self-reported mental health status, among U.S. youth and adults?


## Background and Prior Work

Sleep is a core factor in overall health and daily functioning, and public health guidance typically advises <7 hours/night as insufficient sleep for adults.<sup><a href="#ref1" id="cite1">1</a></sup>

National surveillance data show that insufficient sleep is common in the U.S. and also that insufficient sleep differs across demographic groups.
For example, the CDC analysis of the Behavioral Risk Factor Surveillance System found that only about two-thirds of U.S. adults reported “healthy sleep duration” (≥7 hours), with a lower rate among several racial/ethnic minority groups and among adults without higher education, this shows that there is broader concern for the sleep-health of the overall population rather than individual sleep choices.<sup><a href="#ref2" id="cite2">2</a></sup>
Therefore, sleep duration is a population-level concern.<sup><a href="#ref2" id="cite2b">2</a></sup>

A research paper by the National Library of Medicine has investigated why sleep duration varies by age, sex, race/ethnicity, and socioeconomic status. The medical reviews emphasize that “race/ethnicity” is a social category that can include potential exposures—such as discrimination, poor neighborhood conditions, and working constraints—that ultimately play a role in sleep opportunity and timing.<sup><a href="#ref3" id="cite3">3</a></sup>
Large U.S. survey studies show that disparities stay even after accounting for many variables. For instance, an NHIS trends analysis reported increases in self-reported short sleep in recent years that were especially pronounced among Black and Hispanic adults even after adjusting for demographic and socioeconomic factors, raising concern that widening sleep disparities may lead to other health inequalities over time.<sup><a href="#ref4" id="cite4">4</a></sup>
Longitudinal evidence also suggests that short sleep is more common in lower income/education groups and among some racial/ethnic minority groups, with socioeconomic context and living conditions explaining more of the disparity than individual health behaviors alone.<sup><a href="#ref5" id="cite5">5</a></sup>

Importantly, the consequences of insufficient sleep extend beyond “hours slept” to sleep-related impairment—fatigue, excessive daytime sleepiness, and reduced functioning.
Research that took into account both sleep duration and symptoms found that lack of sleep is associated with higher levels of fatigue and daytime sleepiness, and that both education and race/ethnicity relate to multiple sleep-related symptoms, underscoring that disparities may appear in both sleep quantity and daytime impact.<sup><a href="#ref6" id="cite6">6</a></sup>
Additionally, Mental health is central to this question because it is strongly connected with overall sleep health: evidence supports relationships between sleep disturbances and depression/anxiety, and anxiety can also cause excessive daytime sleepines.<sup><a href="#ref7" id="cite7">7</a></sup>
Also, consistent with that, NHIS-based analysis showed that psychological distress is directly associated with unhealthy sleep duration even after adjusting for socioeconomic factors and health conditions.<sup><a href="#ref8" id="cite8">8</a></sup>
Together, the research shows the importance of our project's focus on estimating how age, sex, race/ethnicity, and socioeconomic status relate to both sleep duration and sleep-related impairment.

### NHANES August 2021 - August 2023 Data ###

This study uses data from the National Health and Nutrition Examination Survey (NHANES), a nationally representative survey of U.S. adults. The analytic dataset was created by merging demographic, sleep questionnaire, and mental questionnaire components. After cleaning and restricting to complete cases, the final sample includes adults with valid data on sleep duration, socioeconomic characteristics, and mental health.

The primary outcome is sleep duration, measured as the number of hours usually slept on weekdays or workdays. Sleep is reported in hours and ranges approximately from 2 to 14 hours in the cleaned dataset. It is top coded at 14 and bottom coded at 2. A value of 2 indicates less than 3 hours of sleep while a value of 14 indicates 14 or more hours of sleep. Short sleep is typically defined as fewer than 7 hours per night and is associated with increased health risks. Mental health was measured using a mental health score, which comprises of nine depression factors with responses scored from 0 (not at all) to 3 (nearly every day). The mental health score ranges from 0 to 27, with higher scores indicating more severe depressive symptoms.

Key predictors include age (years), sex (male/female), race/ethnicity (categorized groups), education level (less than 9th grade to college graduate or above), and income-to-poverty ratio. The income-to-poverty ratio is a continuous measure comparing family income to the poverty threshold. Values below 1 indicate income below the poverty line.

Some limitations should be considered. Sleep and mental health measures are self-reported, so participants may not remember their experiences perfectly or may underreport or overreport symptoms. Additionally, only participants with complete data were included, which could affect results if missing data are not random. Despite these limitations, NHANES is a large and nationally representative dataset that is appropriate for examining differences in sleep duration across demographic and socioeconomic groups.

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 
import pandas as pd

demo = pd.read_sas('data/00-raw/DEMO_L.xpt')
sleep = pd.read_sas('data/00-raw/SLQ_L.xpt')
mental = pd.read_sas('data/00-raw/DPQ_L.xpt')

# Merging demographic, mental health, and sleep data to one Dataframe
df1 = demo.merge(sleep, on="SEQN", how="inner")
df1 = df1.merge(mental, on="SEQN", how="inner")

#Keeping only variables needed
columns_keep = ["SEQN","SLD012","RIDAGEYR","RIAGENDR", "RIDRETH3","INDFMPIR", "DMDEDUC2","DPQ010","DPQ020","DPQ030","DPQ040","DPQ050","DPQ060","DPQ070","DPQ080","DPQ090"]
df1 = df1[columns_keep]


# Pre-clean data summary
print("\nDataset size BEFORE cleaning:")
print(f"Observations (rows): {df1.shape[0]}")
print(f"Variables (cols): {df1.shape[1]}")

missing_summary = pd.DataFrame({
    "missing_count": df1.isna().sum(),
    "missing_percent": (df1.isna().mean() * 100).round(2)
}).sort_values("missing_percent", ascending=False)

print("\nMissingness summary (before cleaning):")
print(missing_summary)

# Flag outliers / suspicious entries 
print("\nPotential outliers / suspicious checks (before cleaning):")

# NHANES codebook notes values typically 3 to 13.5; 2 and 14 are top/bottom codes
sleep_outliers = df1[(df1["SLD012"].notna()) & ((df1["SLD012"] < 2) | (df1["SLD012"] > 14.0))]
print("Sleep outliers count (outside 3–14.0):", sleep_outliers.shape[0])

dpq_cols = ["DPQ010","DPQ020","DPQ030","DPQ040","DPQ050","DPQ060","DPQ070","DPQ080","DPQ090"]
# Mental health total score should be 0–27
for col in dpq_cols:
    print(col, "outliers count:", df1[df1[col] > 3].shape[0])

# Age (0-79); top coded at 80
age_outliers = df1[(df1["RIDAGEYR"].notna()) & ((df1["RIDAGEYR"] < 0))]
print("Age outliers count:", age_outliers.shape[0])

# Income ratio top coded at 5.00
income_outliers = df1[(df1["INDFMPIR"].notna()) & (df1["INDFMPIR"] < 0)]
print("Income ratio outliers count:", income_outliers.shape[0])

#CLEANING

#Add up all the depression data into one score
df1[dpq_cols] = df1[dpq_cols].replace({7: pd.NA, 9: pd.NA})
df1["mental_health_score"] = df1[dpq_cols].sum(axis=1)
df1["mental_health_score"] = df1["mental_health_score"].astype(int)
df1 = df1.drop(columns=dpq_cols)

#Rename columns
df1 = df1.rename(columns={
    "SEQN": "participant_id",
    "SLD012": "sleep_hours",
    "RIDAGEYR": "age",
    "RIAGENDR": "gender",
    "RIDRETH3": "race_ethnicity",
    "INDFMPIR": "income_ratio",
    "DMDEDUC2": "education"})

#Convert age and participant id to integer
df1["participant_id"] = df1["participant_id"].astype(int)
df1["age"] = df1["age"].astype(int)
df1["race_ethnicity"] = df1["race_ethnicity"].astype(int)


# Recode RIDRETH3 (race/ethnicity) into descriptive category labels
# Original NHANES numeric codes are converted to categorical values

race_map = {
    1: "Mexican American",
    2: "Other Hispanic",
    3: "Non-Hispanic White",
    4: "Non-Hispanic Black",
    6: "Non-Hispanic Asian",
    7: "Other/Multi"
}
df1["race_ethnicity"] = df1["race_ethnicity"].map(race_map)

# Keep only valid education levels (remove "Refused" and "Don't Know" responses)
df1 = df1[df1["education"].isin([1,2,3,4,5])]

# Recode DMDEDUC2 (education) into descriptive category labels
# Original NHANES numeric codes are converted to categorical values

education_map = {
    1: "Less than 9th grade",
    2: "9–11th grade",
    3: "High school/GED",
    4: "Some college/AA",
    5: "College graduate or above"
}
df1["education"] = df1["education"].map(education_map)

# Recode RIAGENDR (gender) into descriptive category labels
# Original NHANES numeric codes are converted to categorical values
gender_map = {
    1: "Male",
    2: "Female"
}
df1["gender"] = df1["gender"].map(gender_map)

#Remove rows with NaN values
df1 = df1.dropna()

#After Cleaning Summary

print("\nDataset size AFTER cleaning:")
print(f"Observations (rows): {df1.shape[0]}")
print(f"Variables (cols): {df1.shape[1]}")

missing_summary = pd.DataFrame({
    "missing_count": df1.isna().sum(),
    "missing_percent": (df1.isna().mean() * 100).round(2)
}).sort_values("missing_percent", ascending=False)

print("\nMissingness summary (after cleaning):")
print(missing_summary)

# Flag outliers / suspicious entries 
print("\nPotential outliers / suspicious checks (after cleaning):")

# NHANES codebook notes values typically 3 to 13.5; 2 and 14 are top/bottom codes
sleep_outliers = df1[(df1["sleep_hours"].notna()) & ((df1["sleep_hours"] < 2) | (df1["sleep_hours"] > 14.0))]
print("Sleep outliers count (outside 2–14.0):", sleep_outliers.shape[0])

# Mental health total score should be 0–27
mh_outliers = df1[(df1["mental_health_score"].notna()) & ((df1["mental_health_score"] < 0) | (df1["mental_health_score"] > 27))]
print("Mental health score outliers count:", mh_outliers.shape[0])

# Age (0-79); top coded at 80
age_outliers = df1[(df1["age"].notna()) & ((df1["age"] < 0))]
print("Age outliers count:", age_outliers.shape[0])

# Income ratio top coded at 5.00
income_outliers = df1[(df1["income_ratio"].notna()) & (df1["income_ratio"] < 0)]
print("Income ratio outliers count:", income_outliers.shape[0])

print("Observations:", df1.shape[0])
print("Variables:", df1.shape[1])
df1.to_csv("data/02-processed/nhanes2123_final.csv", index=False)

df1.head()
```

### National Youth Fitness Survey (NNYFS) 2012

The National Youth Fitness Survey (NNYFS) 2012 Demographics dataset contains baseline demographic information for children and teens in the US. NNYFS was conducted by the CDC as a nationally representative survey designed to assess the health, fitness, and nutritional status of U.S. youth. The demographics file provides contextual information that is essential for understanding population characteristics and for analyzing age, sex, and race/ethnicity.

The key variables in this dataset include age, sex, and race/ethnicity. Age is measured in years and represents the participant’s age at the time of the survey. Valid values are non-negative integer under 18. Sex is a categorical variable coded as male or female. Race and ethnicity are recorded using standardized CDC categories, including Mexican American, other Hispanic, non-Hispanic White, non-Hispanic Black, and other racial groups. These demographic variables are not health outcomes themselves, but they are critical explanatory variables that allow us to examine disparities and differences across subpopulations.

There are a few important data quality considerations in this dataset. First, because the data is collected through a large national survey, missingness may arise from nonresponse, skipped questions, or incomplete interviews. Missing demographic values are not expected to be random; for example, race or ethnicity may be missing more frequently among certain age groups. Second, while most variables are categorical, age is a continuous numeric variable and can contain implausible values if survey coding errors occur.  Finally, because the survey focuses exclusively on youth, the results cannot be generalized to adult populations, and any analysis should only be interpreted within the context of children in the United States.

Overall, the dataset provides a clean and structured foundation for data analysis. 

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_raw = pd.read_sas('data/00-raw/Y_DEMO.xpt', format="xport")
df_raw.columns = [c.upper() for c in df_raw.columns]

# ---- Select variables ----
cols = [
    "SEQN",
    "RIDAGEYR",
    "RIAGENDR",
    "RIDRETH1",
]

df = df_raw[cols].copy()
def num(s):
    return pd.to_numeric(s, errors="coerce")

# ---- Clean variables ----
df["age"] = num(df["RIDAGEYR"])
df.loc[df["age"] < 0, "age"] = np.nan

df["sex"] = num(df["RIAGENDR"]).map({
    1: "male",
    2: "female",
})

df["race_ethnicity"] = num(df["RIDRETH1"]).map({
    1: "mexican_american",
    2: "other_hispanic",
    3: "non_hispanic_white",
    4: "non_hispanic_black",
    5: "other_or_multiracial",
})

# ---- Tidy dataset ----
df_clean = df[
    [
        "SEQN",
        "age",
        "sex",
        "race_ethnicity",
    ]
].dropna()

plt.figure()
df_clean["age"].dropna().plot(kind="hist", bins=30)
plt.xlabel("Age (years)")
plt.ylabel("Count")
plt.title("Age distribution")
plt.tight_layout()
plt.show()

plt.figure()
plt.boxplot(df_clean["age"].dropna(), vert=True)
plt.ylabel("Age (years)")
plt.title("Age boxplot (outlier check)")
plt.tight_layout()
plt.show()

print("Observations:", df_clean.shape[0])
print("Variables:", df_clean.shape[1])
df_clean.to_csv('data/02-processed/nnyfs_2012_demo_clean.csv', index=False)

df_clean.head()
```

### OLS Regression
An Ordinary Least Squares (OLS) regression model helps examine whether demographic and socioeconomic factors predict variation in sleep duration while controlling for mental health. The model includes age, gender, race/ethnicity, education level, income-to-poverty ratio, and mental health score as predictors.

```python
import statsmodels.formula.api as smf
model = smf.ols("sleep_hours ~ age + income_ratio + mental_health_score + C(gender) + C(race_ethnicity) + C(education)",data=df1).fit()
print(model.summary())
```

The model is statistically significant (F = 7.33, p < 0.001), indicating that the included variables collectively explain some variation in sleep duration, although the overall explanatory power is relatively low (R² = 0.018).

Among the predictors, mental health score and income-to-poverty ratio show statistically significant relationships with sleep duration. The coefficient for mental health score is negative and significant (β = −0.025, p < 0.001), indicating that higher levels of depressive symptoms are associated with slightly fewer hours of sleep. The income-to-poverty ratio has a small but significant negative association with sleep duration (β = −0.052, p = 0.001), suggesting that individuals with higher income relative to the poverty line report marginally fewer hours of sleep.

Gender also shows a statistically significant effect. The coefficient for males is negative (β = −0.119, p = 0.007), indicating that males report slightly fewer hours of sleep on average compared to females. Other variables, including most race/ethnicity categories, education levels, and age, do not show statistically significant associations with sleep duration at conventional significance levels.
