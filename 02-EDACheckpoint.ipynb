# Research Question

To what extent are age, sex, race/ethnicity, education and socioeconomic status predictive of disparities in sleep duration (hours per night) and sleep-related impairment (self-reported daytime functional and cognitive limitations), after adjusting for self-reported mental health status, among U.S. adults (18+ years old)?


## Background and Prior Work

Sleep is a core factor in overall health and daily functioning, and public health guidance typically advises <7 hours/night as insufficient sleep for adults.<sup><a href="#ref1" id="cite1">1</a></sup>

National surveillance data show that insufficient sleep is common in the U.S. and also that insufficient sleep differs across demographic groups.
For example, the CDC analysis of the Behavioral Risk Factor Surveillance System found that only about two-thirds of U.S. adults reported “healthy sleep duration” (≥7 hours), with a lower rate among several racial/ethnic minority groups and among adults without higher education, this shows that there is broader concern for the sleep-health of the overall population rather than individual sleep choices.<sup><a href="#ref2" id="cite2">2</a></sup>
Therefore, sleep duration is a population-level concern.<sup><a href="#ref2" id="cite2b">2</a></sup>

A research paper by the National Library of Medicine has investigated why sleep duration varies by age, sex, race/ethnicity, and socioeconomic status. The medical reviews emphasize that “race/ethnicity” is a social category that can include potential exposures—such as discrimination, poor neighborhood conditions, and working constraints—that ultimately play a role in sleep opportunity and timing.<sup><a href="#ref3" id="cite3">3</a></sup>
Large U.S. survey studies show that disparities stay even after accounting for many variables. For instance, an NHIS trends analysis reported increases in self-reported short sleep in recent years that were especially pronounced among Black and Hispanic adults even after adjusting for demographic and socioeconomic factors, raising concern that widening sleep disparities may lead to other health inequalities over time.<sup><a href="#ref4" id="cite4">4</a></sup>
Longitudinal evidence also suggests that short sleep is more common in lower income/education groups and among some racial/ethnic minority groups, with socioeconomic context and living conditions explaining more of the disparity than individual health behaviors alone.<sup><a href="#ref5" id="cite5">5</a></sup>

Importantly, the consequences of insufficient sleep extend beyond “hours slept” to sleep-related impairment—fatigue, excessive daytime sleepiness, and reduced functioning.
Research that took into account both sleep duration and symptoms found that lack of sleep is associated with higher levels of fatigue and daytime sleepiness, and that both education and race/ethnicity relate to multiple sleep-related symptoms, underscoring that disparities may appear in both sleep quantity and daytime impact.<sup><a href="#ref6" id="cite6">6</a></sup>
Additionally, Mental health is central to this question because it is strongly connected with overall sleep health: evidence supports relationships between sleep disturbances and depression/anxiety, and anxiety can also cause excessive daytime sleepines.<sup><a href="#ref7" id="cite7">7</a></sup>
Also, consistent with that, NHIS-based analysis showed that psychological distress is directly associated with unhealthy sleep duration even after adjusting for socioeconomic factors and health conditions.<sup><a href="#ref8" id="cite8">8</a></sup>
Together, the research shows the importance of our project's focus on estimating how age, sex, race/ethnicity, and socioeconomic status relate to both sleep duration and sleep-related impairment.

### NHANES August 2021 - August 2023 Data ###

This study uses data from the National Health and Nutrition Examination Survey (NHANES), a nationally representative survey of U.S. adults. The analytic dataset was created by merging demographic, sleep questionnaire, and mental questionnaire components. After cleaning and restricting to complete cases, the final sample includes adults with valid data on sleep duration, socioeconomic characteristics, and mental health.

The primary outcome is sleep duration, measured as the number of hours usually slept on weekdays or workdays. Sleep is reported in hours and ranges approximately from 2 to 14 hours in the cleaned dataset. It is top coded at 14 and bottom coded at 2. A value of 2 indicates less than 3 hours of sleep while a value of 14 indicates 14 or more hours of sleep. Short sleep is typically defined as fewer than 7 hours per night and is associated with increased health risks. Mental health was measured using a mental health score, which comprises of nine depression factors with responses scored from 0 (not at all) to 3 (nearly every day). The mental health score ranges from 0 to 27, with higher scores indicating more severe depressive symptoms.

Key predictors include age (years), sex (male/female), race/ethnicity (categorized groups), education level (less than 9th grade to college graduate or above), and income-to-poverty ratio. The income-to-poverty ratio is a continuous measure comparing family income to the poverty threshold. Values below 1 indicate income below the poverty line.

Some limitations should be considered. Sleep and mental health measures are self-reported, so participants may not remember their experiences perfectly or may underreport or overreport symptoms. Additionally, only participants with complete data were included, which could affect results if missing data are not random. Despite these limitations, NHANES is a large and nationally representative dataset that is appropriate for examining differences in sleep duration across demographic and socioeconomic groups.

```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## FEEL FREE TO ADD MULTIPLE CELLS PER SECTION 
import pandas as pd
import numpy as np

demo = pd.read_sas('data/00-raw/DEMO_L.xpt')
sleep = pd.read_sas('data/00-raw/SLQ_L.xpt')
mental = pd.read_sas('data/00-raw/DPQ_L.xpt')

# Merging demographic, mental health, and sleep data to one Dataframe
df1 = demo.merge(sleep, on="SEQN", how="inner")
df1 = df1.merge(mental, on="SEQN", how="inner")

#Keeping only variables needed
columns_keep = ["SEQN","SLD012","RIDAGEYR","RIAGENDR", "RIDRETH3","INDFMPIR", "DMDEDUC2","DPQ010","DPQ020","DPQ030","DPQ040","DPQ050","DPQ060","DPQ070","DPQ080","DPQ090"]
df1 = df1[columns_keep]


# Pre-clean data summary
print("\nDataset size BEFORE cleaning:")
print(f"Observations (rows): {df1.shape[0]}")
print(f"Variables (cols): {df1.shape[1]}")

missing_summary = pd.DataFrame({
    "missing_count": df1.isna().sum(),
    "missing_percent": (df1.isna().mean() * 100).round(2)
}).sort_values("missing_percent", ascending=False)

print("\nMissingness summary (before cleaning):")
print(missing_summary)

# Flag outliers / suspicious entries 
print("\nPotential outliers / suspicious checks (before cleaning):")

# NHANES codebook notes values typically 3 to 13.5; 2 and 14 are top/bottom codes
sleep_outliers = df1[(df1["SLD012"].notna()) & ((df1["SLD012"] < 2) | (df1["SLD012"] > 14.0))]
print("Sleep outliers count (outside 3–14.0):", sleep_outliers.shape[0])

dpq_cols = ["DPQ010","DPQ020","DPQ030","DPQ040","DPQ050","DPQ060","DPQ070","DPQ080","DPQ090"]
# Mental health total score should be 0–27
for col in dpq_cols:
    print(col, "outliers count:", df1[df1[col] > 3].shape[0])

# Age (0-79); top coded at 80
age_outliers = df1[(df1["RIDAGEYR"].notna()) & ((df1["RIDAGEYR"] < 0))]
print("Age outliers count:", age_outliers.shape[0])

# Income ratio top coded at 5.00
income_outliers = df1[(df1["INDFMPIR"].notna()) & (df1["INDFMPIR"] < 0)]
print("Income ratio outliers count:", income_outliers.shape[0])

#CLEANING

#Add up all the depression data into one score
df1[dpq_cols] = df1[dpq_cols].replace({7: np.nan, 9: np.nan})
df1["mental_health_score"] = df1[dpq_cols].sum(axis=1)
df1["mental_health_score"] = df1["mental_health_score"].astype(int)
df1 = df1.drop(columns=dpq_cols)

#Rename columns
df1 = df1.rename(columns={
    "SEQN": "participant_id",
    "SLD012": "sleep_hours",
    "RIDAGEYR": "age",
    "RIAGENDR": "gender",
    "RIDRETH3": "race_ethnicity",
    "INDFMPIR": "income_ratio",
    "DMDEDUC2": "education"})

#Convert age and participant id to integer
df1["participant_id"] = df1["participant_id"].astype(int)
df1["age"] = df1["age"].astype(int)
df1["race_ethnicity"] = df1["race_ethnicity"].astype(int)


# Recode RIDRETH3 (race/ethnicity) into descriptive category labels
# Original NHANES numeric codes are converted to categorical values

race_map = {
    1: "Mexican American",
    2: "Other Hispanic",
    3: "Non-Hispanic White",
    4: "Non-Hispanic Black",
    6: "Non-Hispanic Asian",
    7: "Other/Multi"
}
df1["race_ethnicity"] = df1["race_ethnicity"].map(race_map)

# Keep only valid education levels (remove "Refused" and "Don't Know" responses)
df1 = df1[df1["education"].isin([1,2,3,4,5])]

# Recode DMDEDUC2 (education) into descriptive category labels
# Original NHANES numeric codes are converted to categorical values

education_map = {
    1: "Less than 9th grade",
    2: "9–11th grade",
    3: "High school/GED",
    4: "Some college/AA",
    5: "College graduate or above"
}
df1["education"] = df1["education"].map(education_map)

# Recode RIAGENDR (gender) into descriptive category labels
# Original NHANES numeric codes are converted to categorical values
gender_map = {
    1: "Male",
    2: "Female"
}
df1["gender"] = df1["gender"].map(gender_map)

#Remove rows with NaN values
df1 = df1.dropna()

#After Cleaning Summary

print("\nDataset size AFTER cleaning:")
print(f"Observations (rows): {df1.shape[0]}")
print(f"Variables (cols): {df1.shape[1]}")

missing_summary = pd.DataFrame({
    "missing_count": df1.isna().sum(),
    "missing_percent": (df1.isna().mean() * 100).round(2)
}).sort_values("missing_percent", ascending=False)

print("\nMissingness summary (after cleaning):")
print(missing_summary)

# Flag outliers / suspicious entries 
print("\nPotential outliers / suspicious checks (after cleaning):")

# NHANES codebook notes values typically 3 to 13.5; 2 and 14 are top/bottom codes
sleep_outliers = df1[(df1["sleep_hours"].notna()) & ((df1["sleep_hours"] < 2) | (df1["sleep_hours"] > 14.0))]
print("Sleep outliers count (outside 2–14.0):", sleep_outliers.shape[0])

# Mental health total score should be 0–27
mh_outliers = df1[(df1["mental_health_score"].notna()) & ((df1["mental_health_score"] < 0) | (df1["mental_health_score"] > 27))]
print("Mental health score outliers count:", mh_outliers.shape[0])

# Age (0-79); top coded at 80
age_outliers = df1[(df1["age"].notna()) & ((df1["age"] < 0))]
print("Age outliers count:", age_outliers.shape[0])

# Income ratio top coded at 5.00
income_outliers = df1[(df1["income_ratio"].notna()) & (df1["income_ratio"] < 0)]
print("Income ratio outliers count:", income_outliers.shape[0])

print("Observations:", df1.shape[0])
print("Variables:", df1.shape[1])
df1.to_csv("data/02-processed/nhanes2123_final.csv", index=False)

df1.head()
```

## Behavioral Risk Factor Surveillance System (BRFSS) 2022

The Behavioral Risk Factor Surveillance System (BRFSS) 2022 dataset contains health-related risk behaviors, chronic health conditions, preventive service use, and sleep information for adults in the United States. BRFSS is conducted annually by the Centers for Disease Control and Prevention (CDC) as a nationally representative, state-based telephone survey. The 2022 dataset provides cross-sectional data used to monitor population health and examine disparities across states and demographic groups.

Key variables include demographic characteristics (age, sex, race/ethnicity, education, income), health behaviors (smoking, alcohol use, physical activity), and health status indicators (e.g., diabetes, hypertension, obesity, general health). Age is recorded in years for adults aged 18 and older, while most other variables are categorical. Importantly, BRFSS 2022 includes a sleep duration measure that asks respondents how many hours of sleep they get in a 24-hour period. This numeric variable (typically 1–24 hours) can be analyzed continuously or categorized to identify short sleep duration (commonly defined as fewer than 7 hours). While it captures sleep quantity, it does not assess sleep quality or diagnosed sleep disorders.

Several data quality considerations apply. Because BRFSS data are self-reported, responses may be affected by recall or social desirability bias. Missing data may result from nonresponse or skipped questions and may not be random. Additionally, BRFSS uses a complex sampling design with weighting, so analyses must apply survey weights to obtain nationally representative estimates. As a cross-sectional survey of noninstitutionalized adults, findings cannot establish causality.

Overall, the dataset provides a structured and comprehensive foundation for analyzing adult health behaviors, chronic disease, sleep duration, and health disparities in the United States.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


brfss_raw = pd.read_sas('data/00-raw/LLCP2022.XPT', format='xport')
brfss_raw.columns = [c.upper() for c in brfss_raw.columns]
print("Total raw observations:", brfss_raw.shape[0])

# ---- Select variables ----
# SLEPTIM1  = sleep hours per night (PRIMARY outcome)
# DIFFALON  = physical/functional limitation (difficulty doing errands alone)
# CNCRDIFF = cognitive limitation (difficulty concentrating or remembering) 
# SEXVAR    = sex
# _AGEG5YR  = age group (5-year bands)
# _IMPRACE  = imputed race/ethnicity
# INCOME3   = household income (SES proxy)
# EDUCA     = education level (SES proxy)
# MENTHLTH  = poor mental health days past 30 (covariate)

cols_wanted = [
    "SLEPTIM1",
    "DIFFALON",
    "CNCRDIFF",
    "SEXVAR",
    "_AGEG5YR",
    "_IMPRACE",
    "INCOME3",
    "EDUCA",
    "MENTHLTH",
]

#keeps existing columns
cols_available = [c for c in cols_wanted if c in brfss_raw.columns]
cols_missing   = [c for c in cols_wanted if c not in brfss_raw.columns]
print("Columns found:  ", cols_available)
print("Columns missing:", cols_missing)

df = brfss_raw[cols_available].copy()

def num(s):
    return pd.to_numeric(s, errors="coerce")

#clean variables
# Sleep hours (1–24 valid; 77=don't know, 99=refused → NaN)
df["sleep_hours"] = num(df["SLEPTIM1"])
df.loc[df["sleep_hours"] > 24, "sleep_hours"] = np.nan
df.loc[df["sleep_hours"] < 1,  "sleep_hours"] = np.nan

# Physical daytime impairment — difficulty doing errands alone
df["functional_impairment"] = num(df["DIFFALON"]).map({1: 1, 2: 0})

# Cognitive impairment — binary (0=no difficulty, 1=any difficulty) to make it easier to comapre w functional impairment
df["cognitive_impairment"] = num(df["CNCRDIFF"]).map({
    1: 0,  # no difficulty
    2: 1,  # some difficulty → impaired
    3: 1,  # a lot of difficulty → impaired
    7: np.nan,
    9: np.nan
})

# Sex (matches teammates' male/female coding)
df["sex"] = num(df["SEXVAR"]).map({
    1: "male",
    2: "female",
})

# Age group — BRFSS uses 5-year bands (matches youth/adult split from teammates)
age_map = {
    1: "18-24",  2: "25-29",  3: "30-34",  4: "35-39",
    5: "40-44",  6: "45-49",  7: "50-54",  8: "55-59",
    9: "60-64", 10: "65-69", 11: "70-74", 12: "75-79",
    13: "80+"
}
df["age_group"] = num(df["_AGEG5YR"]).map(age_map)

# Youth flag (matches teammates' Y_DEMO youth dataset focus)
df["is_youth"] = num(df["_AGEG5YR"]).apply(
    lambda x: True if x == 1 else (False if pd.notna(x) else np.nan)
)

# Race/ethnicity (matches teammates' race_ethnicity naming convention)
race_map = {
    1: "non_hispanic_white",
    2: "non_hispanic_black",
    3: "non_hispanic_asian",
    4: "non_hispanic_other",
    5: "hispanic",
    6: "multiracial",
}
df["race_ethnicity"] = num(df["_IMPRACE"]).map(race_map)

# Income — SES proxy (77=don't know, 99=refused → NaN)
df["income"] = num(df["INCOME3"])
df.loc[df["income"] >= 77, "income"] = np.nan

# Education — SES proxy (9=refused → NaN)
# Matches teammates' education levels
education_map = {
    1: "never_attended",
    2: "elementary",
    3: "some_high_school",
    4: "high_school_grad",
    5: "some_college",
    6: "college_grad",
}
df["education"] = num(df["EDUCA"]).map(education_map)
df.loc[num(df["EDUCA"]) == 9, "education"] = np.nan

# Mental health days — covariate (88=none→0, 77/99=refused/dk → NaN)
df["mental_health_days"] = num(df["MENTHLTH"])
df.loc[df["mental_health_days"] == 88, "mental_health_days"] = 0
df.loc[df["mental_health_days"] >= 77,  "mental_health_days"] = np.nan


#Pre-clean summary
print("\nDataset size BEFORE cleaning:")
print(f"Observations (rows): {df.shape[0]}")
print(f"Variables (cols):    {df.shape[1]}")

missing_summary = pd.DataFrame({
    "missing_count":   df.isna().sum(),
    "missing_percent": (df.isna().mean() * 100).round(2)
}).sort_values("missing_percent", ascending=False)
print("\nMissingness summary (before cleaning):")
print(missing_summary)

#Outlier checks
print("\nOutlier checks (before cleaning):")
sleep_out = df[(df["sleep_hours"].notna()) & 
               ((df["sleep_hours"] < 1) | (df["sleep_hours"] > 24))]
print("Sleep outliers:", sleep_out.shape[0])

mh_out = df[(df["mental_health_days"].notna()) & 
            ((df["mental_health_days"] < 0) | (df["mental_health_days"] > 30))]
print("Mental health days outliers:", mh_out.shape[0])

#Tidy dataset
keep_cols = [
    "sleep_hours",
    "functional_impairment",
    "cognitive_impairment",
    "sex",
    "age_group",
    "is_youth",
    "race_ethnicity",
    "income",
    "education",
    "mental_health_days",
]
keep_cols = [c for c in keep_cols if c in df.columns]
df_clean = df[keep_cols].dropna()

#post-clean summary
print("\nDataset size AFTER cleaning:")
print(f"Observations (rows): {df_clean.shape[0]}")
print(f"Variables (cols):    {df_clean.shape[1]}")

missing_summary_after = pd.DataFrame({
    "missing_count":   df_clean.isna().sum(),
    "missing_percent": (df_clean.isna().mean() * 100).round(2)
}).sort_values("missing_percent", ascending=False)
print("\nMissingness summary (after cleaning):")
print(missing_summary_after)

print("\nObservations:", df_clean.shape[0])
print("Variables:",    df_clean.shape[1])

df_clean.to_csv('data/02-processed/brfss2022_sleep_clean.csv', index=False)
df_clean.head()
```

### OLS Regressions
An Ordinary Least Squares (OLS) regression model helps examine whether demographic and socioeconomic factors predict variation in sleep duration while controlling for mental health. The model includes age, gender, race/ethnicity, education level, income-to-poverty ratio, and mental health score as predictors.

##### NHANEES - OLS Regression

```python
import statsmodels.formula.api as smf
model = smf.ols("sleep_hours ~ age + income_ratio + mental_health_score + C(gender) + C(race_ethnicity) + C(education)",data=df1).fit()
print(model.summary())
```

The model is statistically significant (F = 7.33, p < 0.001), indicating that the included variables collectively explain some variation in sleep duration, although the overall explanatory power is relatively low (R² = 0.018).

Among the predictors, mental health score and income-to-poverty ratio show statistically significant relationships with sleep duration. The coefficient for mental health score is negative and significant (β = −0.025, p < 0.001), indicating that higher levels of depressive symptoms are associated with slightly fewer hours of sleep. The income-to-poverty ratio has a small but significant negative association with sleep duration (β = −0.052, p = 0.001), suggesting that individuals with higher income relative to the poverty line report marginally fewer hours of sleep.

Gender also shows a statistically significant effect. The coefficient for males is negative (β = −0.119, p = 0.007), indicating that males report slightly fewer hours of sleep on average compared to females. Other variables, including most race/ethnicity categories, education levels, and age, do not show statistically significant associations with sleep duration at conventional significance levels.

##### BRFSS - OLS Regression

```python
brfss_model = smf.ols("sleep_hours ~ C(age_group) + C(sex) + C(race_ethnicity) + income + C(education) + mental_health_days", data=df2).fit()

print("=" * 60)
print("OLS REGRESSION — Sleep Hours (BRFSS 2022)")
print("=" * 60)
print(brfss_model.summary())
```

This OLS regression model predicting sleep hours from age, sex, race/ethnicity, income, education, and mental health days was statistically significant overall (F = 31.22, p < 0.001), indicating that the predictors collectively explain a meaningful portion of variance in sleep duration. 

However, the R-squared of 0.042 suggests that the model accounts for only about 4% of the variation in sleep hours, which is consistent with findings from the NHANES dataset and reflects the well-established complexity of sleep as an outcome not fully captured by demographic variables alone. 

The baseline predicted sleep duration was 6.79 hours, representing the reference categories across all predictors. Most age group coefficients were negative relative to the 18-24 reference group, suggesting slightly shorter sleep durations in older age groups, though these differences were not statistically significant. 

Overall, these results suggest that while the demographic and socioeconomic predictors included in the model are jointly significant, no single predictor dominates, which supports the importance of examining the full set of covariates, including mental health, when studying sleep disparities.

### Logic Regressions

Since impairment is a yes/no response, logistic regression is needed to predict the probability that something belongs to one category or the other, keeping predictions bounded between 0 and 1, which makes it appropriate for binary outcomes. The model looks at sleep duration, age, gender, race/ethnicity, education level, income, and mental health score as predictors.

#### BRFSS Logistic Regressions

```python
#functional impairment logistic regression
logit_functional = smf.logit(
    formula="""functional_impairment ~ sleep_hours + C(age_group) + C(sex) + 
               C(race_ethnicity) + income + C(education) + mental_health_days""",
    data=df2
).fit()

print("=" * 60)
print("LOGISTIC REGRESSION — Functional Impairment (BRFSS 2022)")
print("=" * 60)
pd.set_option('display.max_rows', None)
print(logit_functional.summary())

with open('results/logit_functional_summary.txt', 'w') as f:
    f.write(logit_functional.summary2().as_text())

```

The logistic regression predicting functional impairment (difficulty doing errands alone) was statistically significant overall (LLR p < 0.001), with a pseudo R-squared of 0.147, indicating the model explains roughly 15% of the variation in functional impairment (a meaningful fit for a logistic model).

Regarding age, the 40-44 and 65-69 groups showed significantly lower odds of functional impairment compared to the 18-24 reference group (p = 0.038 and p = 0.015 respectively), though most other age groups did not reach significance. Furthermore, sex was a strong and significant predictor: males had significantly lower odds of functional impairment compared to females (coef = -0.44, p < 0.001), suggesting women are more likely to report difficulty with errands.

Race and ethnicity were largely not statistically significant after controlling for other variables, with no racial/ethnic group reaching p < 0.05, though Non-Hispanic Black and multiracial groups showed positive trends. 

Education was a notable predictor given those with elementary education (p = 0.002) and some high school (p = 0.001) had significantly higher odds of functional impairment compared to the college graduate reference group, suggesting a clear SES gradient. 

Income was also highly significant and negative (coef = -0.25, p < 0.001), meaning higher income was strongly associated with lower odds of functional impairment. Mental health days were positively and significantly associated with functional impairment (coef = 0.059, p < 0.001), confirming that poorer mental health is linked to greater daytime functional limitations. Notably, sleep hours alone were not a significant predictor (p = 0.488), suggesting that functional impairment is more strongly driven by socioeconomic and mental health factors than sleep duration alone.

```python
#cog impairment logistic regression
logit_cognitive = smf.logit(
    formula="""cognitive_impairment ~ sleep_hours + C(age_group) + C(sex) + 
               C(race_ethnicity) + income + C(education) + mental_health_days""",
    data=df2
).fit()

print("=" * 60)
print("LOGISTIC REGRESSION — Cognitive Impairment (BRFSS 2022)")
print("=" * 60)
pd.set_option('display.max_rows', None)
print(logit_cognitive.summary())

with open('results/logit_cognitive_summary.txt', 'w') as f:
    f.write(logit_cognitive.summary2().as_text())
```

The logistic regression predicting cognitive impairment (difficulty concentrating or remembering) was statistically significant overall (LLR p < 0.001), though the pseudo R-squared of 0.024 indicates the model explains only about 2% of the variation in cognitive impairment (notably weaker than the functional impairment model). This suggests cognitive impairment is influenced by a broader set of factors not fully captured by the predictors included here.

Age was a significant predictor, with older adults showing progressively higher odds of cognitive impairment compared to the 18-24 reference group. Specifically, the 70-74 (p = 0.032), 75-79 (p = 0.016), and 80+ (p = 0.006) groups all showed significantly elevated odds, reflecting the well-known association between aging and cognitive decline. Unlike functional impairment, sex was not a significant predictor of cognitive impairment (p = 0.686), suggesting cognitive limitations are similarly distributed across males and females in this sample.

Race and ethnicity showed notable disparities. Non-Hispanic White (coef = 0.60, p < 0.001) and Non-Hispanic Other (coef = 0.71, p = 0.001) groups had significantly higher odds of cognitive impairment compared to the Hispanic reference group, while Non-Hispanic Asian respondents had significantly lower odds (coef = -0.75, p = 0.010). 

Education showed limited effects, with only high school graduates having marginally lower odds than college graduates (p = 0.034). Furthermore, income was not a significant predictor (p = 0.582), which contrasts sharply with the functional impairment model where income was one of the strongest predictors. 

Lastly, mental health days remained a significant positive predictor (coef = 0.017, p < 0.001), confirming that poorer mental health is associated with greater cognitive limitations. As with functional impairment, sleep hours alone were not a significant predictor of cognitive impairment (p = 0.492).
